# Fortnite Performance
This is a dataset of fornite player performance

### Import Libraries and Data


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style = 'darkgrid', color_codes = True)
def setplt():
    f, ax = plt.subplots(figsize=(5,3))
    sns.despine(f, left = True, bottom = True)
```

```python
fort_data = pd.read_excel("Fortnite_Statistics.xlsx")
fort_data.head()
```

## Data Overview

```python
fort_data.describe()
```

## Mental State


#### Overall Analysis

```python
# Separate into High and Sober 
high = fort_data[fort_data['Mental State'] == "high"]
sober = fort_data[fort_data['Mental State'] == "sober"]
```

```python
sober.describe()
```

```python
high.describe()
```

#### Violin Plots

```python
setplt()
scat_plt = sns.violinplot(x = 'Mental State', y = "Accuracy", data = fort_data, palette = "Set2")
```

#### Eliminations

```python
setplt()
scat_plt = sns.violinplot(x = 'Mental State', y = "Eliminations", data = fort_data, palette = "Set2")
```

#### Place

```python
setplt() 
v_plt = sns.violinplot(x = 'Mental State', y = "Eliminations", data = fort_data, palette = "Set2")
setplt()
v_plt = sns.violinplot(x = 'Mental State', y = "Placed", data = fort_data, palette = "Set2")
```

Looking at these there seems to be some slight differences in performance depending on mental state.
