## Predicting Mental State with ML

Building a machine-learning algorithm to see if we can predict whether or not a player is high given their game stats.

### Formating data for use with scikit-learn 

```python
# Subsetting for only mental state
mental = fort_data['Mental State']

# Creating targets list
targets = []
[targets.append(1) if x == 'high' else targets.append(sob) for x in mental]

# Creating features data
features = fort_data.drop(['Mental State', 'Date', 'Time of Day'],1)
```

Checking the lenth of each to make sure all data is accounted for

```python
print(len(targets))
print(len(features))
```

### Using the train-test-split model
Builds and training and testing set

```python
from sklearn.cross_validation import train_test_split

features_train, features_test, targets_train, targets_test = train_test_split(features, targets, test_size = 0.4)
```

### Using K-Nearest Neigbors (KNN) Algorithm

#### Importing KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

#### Creating knn classifier with k = 1

```python
knn = KNeighborsClassifier(n_neighbors = 1) # setting neighbors to 1 for now
```

#### Training the model

```python
knn.fit(features_train, targets_train)
```

#### Testing the model

```python
targets_pred = knn.predict(features_test)
print(metrics.accuracy_score(targets_test, targets_pred))
```

#### With k = 5 

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(features_train, targets_train)
targets_pred = knn.predict(features_test)
print(metrics.accuracy_score(targets_test, targets_pred))
```

#### Finding best k-value to use

```python
accuracy_vals = []
k_val = range(1,11)
for x in range(1,11):
    knn = KNeighborsClassifier(n_neighbors=x)
    knn.fit(features_train, targets_train)
    targets_pred = knn.predict(features_test)
    accuracy_vals.append(metrics.accuracy_score(targets_test, targets_pred))
    
d = {"K-Value": k_val, "Accuracy": accuracy_vals}
ac_df = pd.DataFrame(data = d)

setplt()
lnplt = sns.lineplot(x = "K-Value", y = "Accuracy", data = ac_df)
```

```python
# Looking at k = 2
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(features_train, targets_train)
targets_pred = knn.predict(features_test)
print(metrics.accuracy_score(targets_test, targets_pred))
```

### Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression()
reg.fit(features_train, targets_train)
reg_pred = reg.predict(features_test)
print(metrics.accuracy_score(targets_test, targets_pred))
```

### Using Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 100)
rf.fit(features_train, targets_train)
rf_pred = rf.predict(features_test)
print(metrics.accuracy_score(targets_test, rf_pred))
```

### Support Vector Classifier (SVC)

```python
from sklearn import svm, model_selection

sv = svm.SVC(gamma = 0.0001, C = 100)
sv.fit(features_train, targets_train)
sv_pred = sv.predict(features_test)
print(metrics.accuracy_score(targets_test, sv_pred))
```

#### Choosing best gamma and C values

```python
param_grid = {"C": [1,10,100], "gamma": [0.1, 0.01, 0.001, 0.005, 0.0001]}
gridsearch = model_selection.GridSearchCV(svm.SVC(), param_grid)
gridsearch.fit(features_test, targets_test)
print(gridsearch.best_params_)
```

```python
sv = svm.SVC(gamma = 0.1, C = 10)
sv.fit(features_train, targets_train)
sv_pred = sv.predict(features_test)
print(metrics.accuracy_score(targets_test, sv_pred))
```
